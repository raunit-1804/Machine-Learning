```python
import numpy as np
```

```python
def costF(X_train, Y_train, coeff):
    M = len(X_train)
    cost = 0
    N = X_train.shape[1]
    for i in range(M):
        pred = 0
        for j in range(N):
            x = X_train[i][j]
            pred += x*coeff[j]
        pred += coeff[N]
        y = Y_train[i]
        cost += (1/M)*((y - pred) ** 2) 
    print("Cost: ", cost)
```

```python
def stepGradient(X_train, Y_train, coeff, alpha):
    coeff_slope = []
    M = len(X_train)
    N = X_train.shape[1]
    coeff_slope = np.zeros(N + 1)
    for i in range(M):
        pred = 0
        for j in range(N):
            x = X_train[i][j]
            pred += x*coeff[j]
        pred += coeff[N]
        for j in range(N):
            x = X_train[i][j]
            y = Y_train[i]
            coeff_slope[j] += (-2/M)*(y - pred)*x
        coeff_slope[N] = (-2/M)*(y - pred)
    for i in range(N + 1):
        coeff[i] = coeff[i] - alpha*coeff_slope[i]
    return coeff
```

```python
def gd(X_train, Y_train, alpha, numIterations):
    M = len(X_train)
    N = X_train.shape[1]
    coeff = np.zeros(N + 1)
    for i in range(numIterations):
        coeff = stepGradient(X_train, Y_train, coeff, alpha)
        costF(X_train, Y_train, coeff)
    return coeff    
```

```python
def run():
    import numpy as np
    from sklearn import preprocessing
    data = np.genfromtxt("/Users/raunitsingh/Downloads/0000000000002417_training_boston_x_y_train.csv", delimiter = ",")
    X_trainRaw = data[:, 0:13]
    Y_train = data[:, 13]
    scaler = preprocessing.StandardScaler().fit(X_trainRaw)
    X_train = scaler.transform(X_trainRaw)
    #scaler2 = preprocessing.StandardScaler().fit(Y_trainRaw)
    #Y_train = scaler2.transform(Y_trainRaw)
    
    alpha = 0.16
    numIterations = 1110
    coeff = gd(X_train, Y_train, alpha, numIterations)
    Y_train
    return coeff, scaler
```

```python
def prediction(X_test, coeff):
    N = X_test.shape[1]
    coeffCopy = np.zeros(N)
    for i in range(N):
        coeffCopy[i] = coeff[i]
    coeffMatrix = np.asmatrix(coeffCopy).T
    xMatrix = np.asmatrix(X_test)
    y_pred = np.squeeze(np.asarray(xMatrix*coeffMatrix))
    y_pred = y_pred + coeff[N]
    return y_pred
```
