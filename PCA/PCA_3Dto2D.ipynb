```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
np.random.seed(2343243)
```

```python
mean_vec1 = np.array([0,0,0])
cov_mat1 = np.array([[1,0,0], [0,1,0], [0,0,1]])
class1 = np.random.multivariate_normal(mean_vec1, cov_mat1, 100)
class1
```

```python
mean_vec2 = np.array([1,1,1])
cov_mat2 = np.array([[1,0,0], [0,1,0], [0,0,1]])
class2 = np.random.multivariate_normal(mean_vec2, cov_mat2, 100)
class2
```

```python
from mpl_toolkits.mplot3d import Axes3D, proj3d
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection = '3d')
ax.plot(class1[:, 0], class1[:, 1], class1[:, 2], 'o')
ax.plot(class2[:, 0], class2[:, 1], class2[:, 2], '^')
plt.show()
```

```python
all_data = np.concatenate((class1, class2))
```

```python
pca = PCA(n_components = 2)
transformed_data = pca.fit_transform(all_data)
transformed_data
```

```python
pca.components_
```

```python
plt.plot(transformed_data[0:100, 0], transformed_data[0:100, 1], "o")
plt.plot(transformed_data[100:200, 0], transformed_data[100:200, 1], "^")
plt.show()
```

```python
X_approx = pca.inverse_transform(transformed_data)
X_approx
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(111, projection = '3d')
ax.plot(X_approx[0:100, 0], X_approx[0:100, 1], X_approx[0:100, 2], 'o')
ax.plot(X_approx[100:200, 0], X_approx[100:200, 1], X_approx[100:200, 2], '^')
plt.show()
```

```python
all_data_t = all_data.T
cov_matrix = np.cov(all_data_t)
cov_matrix
```

```python
eigValue_correspondingVector = []
eig_val, eig_vectors = np.linalg.eig(cov_matrix)
for i in range(len(eig_val)):
    eig_vec = eig_vectors[:, i]
    eigValue_correspondingVector.append((eig_val[i], eig_vec))
eigValue_correspondingVector.sort(reverse = True)
eigValue_correspondingVector
#Now we can choose the top 'k' eigen vectors as our new directions
```
