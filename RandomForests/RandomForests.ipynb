```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv("/Users/raunitsingh/Downloads/0000000000002429_training_titanic_x_y_train.csv")
print(titanic.shape)
titanic.head()
```

```python
from sklearn.model_selection import train_test_split
X = titanic.iloc[:, 0 : titanic.shape[1] - 1]
Y = titanic.iloc[:, -1]

titanic.head()
```

```python
def fun(s):
    if(s == "male"):
        return 0
    else:
        return 1
X["Sex"] = X["Sex"].apply(fun)

X.head()
```

```python
print(X.isnull().sum())
X = X.fillna(X.median())
X.head()
del X["Cabin"]
del X["Name"]
del X["Ticket"]
print(X.isnull().sum())
```

```python
def embarked(s):
    if(s == "C"):
        return 0
    elif(s == "S"):
        return 1
    else:
        return 2
X["Embarked"] = X["Embarked"].apply(embarked)
print(X["Embarked"])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf.fit(X_train, Y_train)
Y_pred_1 = clf.predict(X_train) 
Y_pred_2 = clf.predict(X_test)
clf.score(X_train, Y_train), clf.score(X_test, Y_test) 
```

```python
from sklearn.ensemble import RandomForestClassifier
clf_Random = RandomForestClassifier(max_depth = 6, random_state = 1)
clf_Random.fit(X_train, Y_train)
Y_pred_1 = clf_Random.predict(X_train)
Y_pred_2 = clf_Random.predict(X_test)
clf_Random.score(X_train, Y_train), clf_Random.score(X_test, Y_test) 
```
